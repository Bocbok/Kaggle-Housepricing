# Create ML set

```python
y = df_train['SalePrice']
X = df_train.drop(['SalePrice'], axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
validation_set_size = round(0.2 * len(X))

X_train = X.iloc[validation_set_size:]
X_validation = X.iloc[:validation_set_size]

y_train = y.iloc[validation_set_size:]
y_validation = y.iloc[:validation_set_size]
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

sc = MinMaxScaler()

X_sc = pd.DataFrame(sc.fit_transform(X.copy()))
df_test_sc = pd.DataFrame(sc.transform(df_test.copy()))

X_sc_train = X_sc.iloc[validation_set_size:]
X_sc_validation = X_sc.iloc[:validation_set_size]
```

# Regression Model

```python
def rmsle(predictions, targets):
    return np.sqrt(((np.log(predictions) - np.log(targets)) ** 2).mean())
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(max_iter=10000)
lasso.fit(X_train, y_train)
predictions = lasso.predict(X_validation)

print(rmsle(predictions, y_validation))
```

```python
lasso2 = Lasso(max_iter=100000, normalize=True)
lasso2.fit(X_train, y_train)
predictions = lasso2.predict(X_validation)

print(rmsle(predictions, y_validation))
```

```python
lasso3 = Lasso(max_iter=10000, normalize=True)
lasso3.fit(X_sc_train, y_train)
predictions = lasso3.predict(X_sc_validation)

print(rmsle(predictions, y_validation))
```

# Log transform

```python
df_train_log = df_train.copy()
df_train_log['SalePrice_Log'] = np.log(df_train_log['SalePrice'])
df_train_log = df_train_log.drop(['SalePrice'], axis=1)
```

```python
y_log = df_train_log['SalePrice_Log']
X = df_train_log.drop(['SalePrice_Log'], axis=1)

X.head()
```

```python
y_log.head()
```

```python
X_train = X.iloc[validation_set_size:]
X_validation = X.iloc[:validation_set_size]

y_log_train = y_log.iloc[validation_set_size:]
y_log_validation = y_log.iloc[:validation_set_size]
```

```python
from sklearn.metrics import mean_squared_error

def rmse(predictions, targets):
    return np.sqrt(mean_squared_error(targets, predictions))
```

```python
lasso4 = Lasso()
lasso4.fit(X_train, y_log_train)
predictions = lasso4.predict(X_validation)
print(rmse(y_log_validation, predictions))
```

# Lasso with cross validation

```python
from sklearn.linear_model import LassoCV

reg = LassoCV(cv=5, normalize=True).fit(X_train, y_log_train)
print(reg.score(X_train, y_log_train))
predictions = reg.predict(X_validation)
print(rmse(y_log_validation, predictions))
```

```python
sub = pd.DataFrame()
sub['Id'] = df_test.reset_index()['Id']
sub['SalePrice'] = np.exp(reg.predict(df_test))

sub.to_csv('submission.csv',index=False)
```
